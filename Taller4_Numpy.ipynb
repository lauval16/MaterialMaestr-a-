La regresión lineal es una técnica estadística utilizada para modelar la relación entre una variable dependiente "y" y una o más variables independientes "x".En este taller, resolveremos el problema de regresión lineal simple utilizando la fórmula matricial de minimos cuadrados :

                                 β=(X^T X)^{-1} X^T y
donde:

• X: es la matriz de diseño

• y: es el vector de valores observados

• β: es el vector de coeficientes
                               

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Datos
x=np.array([1,2,3,4,5]) #Horas de estudio
y=np.array([1,3,6,7,10]) #Calificación

#Matriz de diseño
X=np.vstack([np.ones_like(x),x]).T

#Coeficientes β 
beta=np.dot(np.linalg.inv(np.dot(X.T,X)),np.dot(X.T,y))

# Coeficientes [b,m]
intercepto=beta[0]
pendiente=beta[1]
print(f"Coeficiente(intercepto) b:{intercepto}")
print(f"Pendiente m:{pendiente}")

#Predicción con la recta de regresión
y_pred=np.dot(X,beta)

#Gráfica de los datos originales (puntos)
plt.scatter(x,y,color="blue",label="Datos",marker="o")

#Gráfica de la linea de regresión 
plt.plot(x,y_pred,color="red",label=f"Regresión Lineal: y= {pendiente:.2f}x + {intercepto: .2f}",linewidth=2)

#Etiquetas y titulo
plt.xlabel("Horas de estudio")
plt.ylabel("Calificación")
plt.title("Regresión Lineal: Horas de estudio vs Calificación")

#Mostrar leyenda
plt.legend()
plt.grid(True)
```
